### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

for nombre, total in ingresos.items():
    print(f"{nombre}: ${total:,.2f}")
```

```python
import matplotlib.pyplot as plt

plt.bar(ingresos.keys(), ingresos.values(), color='skyblue')
plt.title("Ingresos totales por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos en pesos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Contar cuántas veces se vendió cada categoría por tienda
categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Mostrar resultados para cada tienda
print("Ventas por categoría - Tienda 1:\n", categorias_tienda1, "\n")
print("Ventas por categoría - Tienda 2:\n", categorias_tienda2, "\n")
print("Ventas por categoría - Tienda 3:\n", categorias_tienda3, "\n")
print("Ventas por categoría - Tienda 4:\n", categorias_tienda4, "\n")
```

```python
# Contar ventas por categoría de cada tienda
categorias_tienda1 = tienda["Categoría del Producto"].value_counts()
categorias_tienda2 = tienda2["Categoría del Producto"].value_counts()
categorias_tienda3 = tienda3["Categoría del Producto"].value_counts()
categorias_tienda4 = tienda4["Categoría del Producto"].value_counts()

# Crear figura con 4 subplots en una fila
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Tienda 1
categorias_tienda1.plot.pie(
    ax=axes[0],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
axes[0].set_title("Tienda 1")
axes[0].set_ylabel("")

# Tienda 2
categorias_tienda2.plot.pie(
    ax=axes[1],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
axes[1].set_title("Tienda 2")
axes[1].set_ylabel("")

# Tienda 3
categorias_tienda3.plot.pie(
    ax=axes[2],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
axes[2].set_title("Tienda 3")
axes[2].set_ylabel("")

# Tienda 4
categorias_tienda4.plot.pie(
    ax=axes[3],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'black'}
)
axes[3].set_title("Tienda 4")
axes[3].set_ylabel("")

plt.suptitle("Distribución de ventas por categoría - Todas las tiendas", fontsize=19)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()

```

# 3. Calificación promedio de la tienda


```python
promedios = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

# Mostrar resultados
for nombre, promedio in promedios.items():
    print(f"{nombre}: {promedio:.2f} estrellas")
```

```python
import matplotlib.pyplot as plt

plt.bar(promedios.keys(), promedios.values(), color='orange')
plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Promedio de estrellas")
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Obtener conteos por producto
ventas = {
    "Tienda 1 - Top 5 Mas vendidos": tienda["Producto"].value_counts().head(5),
    "Tienda 1 - 5 Menos vendidos": tienda["Producto"].value_counts().tail(5),
    "Tienda 2 - Top 5 Mas vendidos": tienda2["Producto"].value_counts().head(5),
    "Tienda 2 - 5 Menos vendidos": tienda2["Producto"].value_counts().tail(5),
    "Tienda 3 - Top 5 Mas vendidos": tienda3["Producto"].value_counts().head(5),
    "Tienda 3 - 5 Menos vendidos": tienda3["Producto"].value_counts().tail(5),
    "Tienda 4 - Top 5 Mas vendidos": tienda4["Producto"].value_counts().head(5),
    "Tienda 4 - 5 Menos vendidos": tienda4["Producto"].value_counts().tail(5),
}

# Lista de títulos para la navegación
titulos = list(ventas.keys())

# Widget de índice
indice = widgets.IntSlider(value=0, min=0, max=len(titulos)-1, step=1, description='Gráfico')

# Botones de navegación
btn_prev = widgets.Button(description="⏪ Anterior")
btn_next = widgets.Button(description="Siguiente ⏩")

# Función para actualizar gráfico
def mostrar_grafico(index):
    clear_output(wait=True)

    titulo = titulos[index]
    data = ventas[titulo]

    fig, ax = plt.subplots(figsize=(8, 6))
    barras = ax.bar(data.index, data.values, color='green' if 'Top' in titulo else 'red')

    for barra in barras:
        altura = barra.get_height()
        ax.text(barra.get_x() + barra.get_width()/2, altura/2, f"{int(altura)}", ha='center', va='center', color='white', fontsize=10, fontweight='bold')

    ax.set_title(titulo)
    ax.set_ylabel("Cantidad de ventas")
    ax.set_xticklabels(data.index, rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    display(widgets.HBox([btn_prev, btn_next]))
    display(indice)

# Acciones de los botones
def al_hacer_click_anterior(b):
    if indice.value > 0:
        indice.value -= 1

def al_hacer_click_siguiente(b):
    if indice.value < len(titulos) - 1:
        indice.value += 1

# Conectar acciones
btn_prev.on_click(al_hacer_click_anterior)
btn_next.on_click(al_hacer_click_siguiente)
indice.observe(lambda change: mostrar_grafico(change.new), names='value')

# Mostrar primer gráfico
mostrar_grafico(indice.value)
```

# 5. Envío promedio por tienda

```python
envios_promedio = {
    "Tienda 1": tienda["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

# Mostrar resultados con formato bonito
for nombre, costo in envios_promedio.items():
    print(f"{nombre}: ${costo:.2f} MXN")
```

```python
import matplotlib.pyplot as plt

plt.bar(envios_promedio.keys(), envios_promedio.values(), color='purple')
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Promedio en MXN")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

---

### 💰 Ingresos totales

La tienda con mayor facturación fue **Tienda 1**, mientras que la **Tienda 4** presentó el ingreso más bajo. Aunque la diferencia no es abismal, sí marca una tendencia de bajo rendimiento financiero en esta última.

---

### 📦 Ventas por categoría

Todas las tiendas presentan variedad en sus ventas por categoría. Sin embargo, **Tienda 4** mostró menor participación en categorías de alta rotación como Electrónicos o Muebles, lo que puede estar afectando su volumen de ventas.

---

### ⭐ Calificación promedio

La **Tienda 3** obtuvo la mejor calificación promedio, seguida por la Tienda 1. **Tienda 4** tuvo una de las calificaciones más bajas, lo que podría ser reflejo de problemas en la experiencia del cliente o calidad del producto.

---

### 🛍️ Productos más y menos vendidos

- **Tienda 1**: Más vendido – *Microondas*, menos vendido – *Auriculares con micrófono*  
- **Tienda 2**: Más vendido – *Iniciando en programación*, menos vendido – *Juego de mesa*  
- **Tienda 3**: Más vendido – *Kit de bancas*, menos vendido – *Bloques de construcción*  
- **Tienda 4**: Más vendido – *Cama box*, menos vendido – *Guitarra eléctrica*  

Se observa que Tienda 4 vende productos de mayor tamaño o valor, pero de menor rotación, lo que puede ralentizar el flujo de efectivo.

---

### 🚚 Costo de envío promedio

La **Tienda 4** mostró uno de los **costos de envío más altos**. Esto puede deberse a la naturaleza de sus productos (más grandes, pesados o frágiles), y puede impactar negativamente la satisfacción del cliente o los márgenes operativos.
